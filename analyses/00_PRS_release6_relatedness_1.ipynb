## Notebook Overview
- Estimate relatedness (filter for second-degree relationship) estimated by using a --king-cutoff 0.088 and remove related individuals

```bash
## Load packages
module load plink/2.0-alpha
```

```bash
cd ${WORK_DIR}/raw_genotypes/

# List of ancestries
ANCESTRIES=("AAC" "AFR" "AJ" "AMR" "EAS" "EUR" "CAS")

# Iterate over ancestries
for ANCESTRY in "${ANCESTRIES[@]}"; do
    plink2 \
    --pfile "${ANCESTRY}"/"${ANCESTRY}_release6" \
    --king-cutoff 0.088 \
    --make-bed \
    --out "${ANCESTRY}"/"unrelated_${ANCESTRY}_release6"
done
```

```bash
## Format files
# List of ancestries
ANCESTRIES=("AAC" "AFR" "AJ" "AMR" "EAS" "EUR" "CAS")

# Iterate over ancestries
for ANCESTRY in "${ANCESTRIES[@]}"; do
    cd "$ANCESTRY"    
    awk '{print $1, $1}' "unrelated_${ANCESTRY}_release6.king.cutoff.out.id" > "related_${ANCESTRY}"
    wc -l "related_${ANCESTRY}"
 cd ..  # Return to the original directory if needed
done
```

```bash
## UNRELATED OVERVIEW
# List of ancestries
ANCESTRIES=("AAC" "AFR" "AJ" "AMR" "EAS" "EUR" "CAS")

# Iterate over ancestries
for ANCESTRY in "${ANCESTRIES[@]}"; do
    cd "$ANCESTRY"    
    awk '$6== 2' "unrelated_${ANCESTRY}_release6.fam" > ${ANCESTRY}_cases
    wc -l ${ANCESTRY}_cases
    awk '$6== 1' "unrelated_${ANCESTRY}_release6.fam" > ${ANCESTRY}_controls
    wc -l ${ANCESTRY}_controls
 cd ..  # Return to the original directory if needed
done
```
