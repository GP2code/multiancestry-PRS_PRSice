```python
df_z = pd.read_csv("thresholds.csv")
df_z.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Assuming df_z is your DataFrame

# List of column names
cols = list(df_z.columns)[1:]

# Specify the desired color values
custom_colors = [0, 1, 3, 5, 7, 9, 10, 12, 14, 16, 18, 20, 22, 24, 26]

# Create a custom colormap based on the specified colors
custom_cmap = ListedColormap(sns.color_palette("Spectral_r", n_colors=len(custom_colors)))

# Create heatmap with custom color palette
import matplotlib as mpl
norm = mpl.colors.Normalize(vmin=0, vmax=30)

# Create heatmap with custom color palette
ax = sns.heatmap(df_z[cols], cmap=custom_cmap, yticklabels=df_z['ancestry'], norm=norm)
ax.set_ylabel('ancestry')

# Save the plot
plt.savefig('heatmap_release6_PRSice.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Assuming df_z is your DataFrame

# List of column names
cols = list(df_z.columns)[1:]

# Specify the desired color values
custom_colors = [0, 1, 3, 5, 7, 9, 10, 12, 14, 16, 18, 20, 22, 24, 26]

# Create a custom colormap based on the specified colors
custom_cmap = ListedColormap(sns.color_palette("Spectral_r", n_colors=len(custom_colors)))

# Create heatmap with custom color palette
import matplotlib as mpl
norm = mpl.colors.Normalize(vmin=0, vmax=30)
ax = sns.heatmap(df_z[cols], cmap=custom_cmap, yticklabels=df_z['Ancestry'], norm=norm)
ax.set_ylabel('Ancestry')

# Save the plot
plt.savefig('heatmap_release6_conventional.png', dpi=300, bbox_inches='tight')
plt.show()
```
